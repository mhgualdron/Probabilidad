## Distribución normal teórica
Es al distribución más famosa de toda slas avriables continuas en en el universo de las matemáticas.

$$P(X) = \frac{1}{\sigma \sqrt{2 \pi}} \exp{\left[-\frac{1}{2}\left(\frac{X-\mu}{\sigma} \right)^2 \right]}$$

* $\mu$: media de la distribución
* $\sigma$: desviación estándar de la distribución

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# definimos nuestra distribución gaussiana
def gaussian(x, mu, sigma):
      return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```python
x = np.arange(-4,4,0.1)
y = gaussian(x, 0, 1)

plt.plot(x, y)
```

```python
# usando scipy
dist = norm(0, 1)
x = np.arange(-4,4,0.1)
#aplicando la función de densidad de probabilidad .pdf a la norm
y = [dist.pdf(value) for value in x]
plt.plot(x, y)
```

```python
# calculando la distribución acumulada
dist = norm(0, 1)
x = np.arange(-4,4,0.1)
#Como todas las distribuciones, la normal tiene la cummulative densitiy function

y = [dist.cdf(value) for value in x]
plt.plot(x, y)
```

## Distribución normal (gausiana) a partir de los datos

El [DataSet](https://seattlecentral.edu/qelp/sets/057/057.html) a trabajar corresponde con la medición que se tiene en las alas de las moscas caseras tras un estudio.

```python
df = pd.read_excel('s057.xls')
display(df)
```

```python
arr = df['Normally Distributed Housefly Wing Lengths'].values[4:]
values, dist = np.unique(arr, return_counts=True)
print(values)
plt.bar(values, dist)
```

```python
# estimación de la distribución de probabilidad
mu = arr.mean()

#distribución teórica
sigma = arr.std()
dist = norm(mu, sigma)
x = np.arange(30,60,0.1)
y = [dist.pdf(value) for value in x]
plt.plot(x, y)

# datos
values, dist = np.unique(arr, return_counts=True)
plt.bar(values, dist/len(arr))
```
